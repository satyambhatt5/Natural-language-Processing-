```
import string

```

```
puch_table=str.maketrans("","",string.punctuation)
```

```
txt.translate(puch_table)
```

```
#remove the all \n

cleaned_txt=txt.translate(puch_table).replace('\n',' ')
cleaned_txt
```

```
#cleaned data 


cleaned_low=word_tokenize(cleaned_txt)


len(cleaned_low)
```
