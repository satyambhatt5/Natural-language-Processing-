```
#STEMMING IS THE WORD THAT process the normalization which involve the reducing the word root/base from the derivative affix or derive word 
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize,word_tokenize
ps=PorterStemmer()
```

```
example_word= ['like','likely','liking','liked','unlike']
```
